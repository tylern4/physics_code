```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.ndimage import gaussian_filter
from scipy import interpolate
```

```python
data = pd.read_csv('Amdahl.dat',names=['threads','time'])
data1 = data[:48].reset_index()
data2 = data[48:96].reset_index()
data3 = data[96:144].reset_index()
data4 = data[144:].reset_index()


avg = (data1+data2)/2.0




data3.head()
```

```python
plt.figure(figsize=(16,9))


plt.plot(avg['threads'], avg['time'][0]/avg['time'], label='speedup')
plt.scatter(avg['threads'], avg['time'][0]/avg['time'])
plt.plot(avg['threads'], avg['threads'], label='linear speedup')
plt.ylim(0,30)

plt.title("Speedup of Analysis Program", fontsize=24)
# Intel(R) Xeon(R) Platinum 8260 CPU @ 2.40GHz
plt.xlabel("# Threads", fontsize=24)
plt.ylabel("Speedup (S)", fontsize=24)

plt.tick_params(axis='both', which='major', labelsize=20)

plt.text(40, 9.0, 
         "Intel Xeon Platinum 8260 CPU\n24 core    48 threads    2.40GHz", size=20, ha="center", va="center")
plt.text(40, 5.0, 
         "$S = \dfrac{time_{single}}{time_{\# threads}}$", size=20, ha="center", va="center")

plt.legend(fontsize=20)


plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.plot(data1['threads'], data1['time'])
plt.plot(data2['threads'], data2['time'])
plt.plot(data3['threads'], data3['time'])
plt.plot(data4['threads'], data4['time'])

plt.yscale("log")
plt.show()
```
